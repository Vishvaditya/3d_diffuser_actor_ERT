```python
import os
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
RESULT_PATH = r"/data/KonstantinosKarydis/Vishva/projects/3d_diffuser_actor_ERT/train_logs/Planner_Calvin/pretrained/eval_logs/"
```

```python
def openResults(seed=0, annot=0, mode='ert'):
    df = pd.read_table(os.path.join(RESULT_PATH, f"result_{mode}_seed_{seed}_annot_{annot}.txt"),
                       sep='|',
                       header=None,
                       names=['index', 'task', 'annotation', 'success'])
    df.drop(columns=['index'],
            inplace=True)
    df['seed'] = seed
    df['mode'] = mode
    df['annot_num'] = annot
    df = df[['task', 'mode', 'seed', 'annot_num', 'annotation', 'success']]
    return df
```

# ERT

```python
df = pd.DataFrame()
for i in range(5):
    temp_df = openResults(0, i)
    df = pd.concat([df, temp_df])

df
```

```python
for seed in [17,23,42]:
    temp_df = openResults(seed, 0)
    df = pd.concat([df, temp_df])

df
```
